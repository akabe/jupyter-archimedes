# jupyter-archimedes

[jupyter-archimedes](https://github.com/akabe/jupyter-archimedes) is a Jupyter-friendly 2D plotting library based on [Archimedes](http://archimedes.forge.ocamlcore.org/). jupyter-archimedes adds [OCaml Jupyter](akabe.github.io/ocaml-jupyter/) backend to Archimedes, a 2D plotting library written in OCaml. You can plot and embed figures on Jupyter.

```ocaml
#require "jupyter-archimedes"
```

This figure shows $y = sin(x)$ where $x \in [0, 10]$.

```ocaml
let vp = A.init ~w:560. ~h:260. ["jupyter"] in
A.Axes.box vp ;
A.set_color vp A.Color.red ;
A.fx vp sin 0.0 10.0 ;
A.close vp
```

We show a little complex example "random walk".

```ocaml
let rec walk acc p i =
  if i = 0 then List.rev acc else begin
    let q = if Random.bool () then succ p else pred p in
    walk (q :: acc) q (pred i)
  end in
let vp = A.init ~w:560. ~h:260. ["jupyter"] in
A.Axes.box vp ;
A.set_color vp A.Color.red ;
A.List.y vp ~style:`Lines (List.map float @@ walk [] 0 1000) ; (* Plot a random walk of 1000 points. *)
A.close vp
```
